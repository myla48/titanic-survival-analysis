Part 2: Real Dataset — Titanic

Load and Explore Dataset

```python
import pandas as pd

# Load dataset
titanic_df = pd.read_csv("/kaggle/input/datasets/ruhamazerihun/train-dataset/train.csv")

# Step 1: Exploration
print("=== Head ===")
display(titanic_df.head())

print("=== Info ===")
titanic_df.info()

print("=== Describe ===")
display(titanic_df.describe())
```

Data Cleaning

```python
# Step 2: Data Cleaning
titanic_df = titanic_df.copy()  # avoid chained assignment issues
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].median())
titanic_df['Embarked'] = titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0])
titanic_df = titanic_df.drop(columns=['Cabin'])
titanic_df = titanic_df.drop_duplicates()

# Check for remaining missing values
titanic_df.isnull().sum()
```

Data Analysis

```python
# Step 3: Analysis
survival_by_gender = titanic_df.groupby('Sex', observed=True)['Survived'].mean()
survival_by_class = titanic_df.groupby('Pclass', observed=True)['Survived'].mean()
average_age_per_class = titanic_df.groupby('Pclass', observed=True)['Age'].mean()

# Create Age Groups
bins = [0, 12, 18, 35, 60, 80]
labels = ['Child','Teen','Young Adult','Adult','Senior']
titanic_df['AgeGroup'] = pd.cut(titanic_df['Age'], bins=bins, labels=labels)
survival_by_age_group = titanic_df.groupby('AgeGroup', observed=True)['Survived'].mean()

# Display analysis
print("Survival rate by gender:\n", survival_by_gender)
print("\nSurvival rate by class:\n", survival_by_class)
print("\nAverage age per class:\n", average_age_per_class)
print("\nSurvival rate by age group:\n", survival_by_age_group)
```

Filtering

```python
# Step 4: Filtering
female_survived = titanic_df[(titanic_df['Sex']=='female') & (titanic_df['Survived']==1)]
children_survived = titanic_df[(titanic_df['Age'] <= 12) & (titanic_df['Survived']==1)]
first_class_high_survival = titanic_df[(titanic_df['Pclass']==1) & (titanic_df['Survived']==1)]

print(f"Females who survived: {len(female_survived)}")
print(f"Children who survived: {len(children_survived)}")
print(f"1st class passengers who survived: {len(first_class_high_survival)}")
```

Insights

```python
# Step 5: Insights
most_likely_to_survive_gender = 'female' if survival_by_gender['female'] > survival_by_gender['male'] else 'male'
print(f"Most likely to survive by gender: {most_likely_to_survive_gender}")
print("Class effect: Yes, higher class increased survival probability.")
children_prioritized = "Yes, children had higher survival rates." if survival_by_age_group['Child'] > survival_by_age_group.mean() else "No."
print(f"Children prioritized? {children_prioritized}")
print("Highest survival combination: Female in 1st class, young adult or child age group.")
```
