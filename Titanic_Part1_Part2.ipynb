## Part 2: Real Dataset — Titanic

```python
import pandas as pd

# -----------------------------
# Load Titanic Dataset
# -----------------------------
titanic_df = pd.read_csv('/content/Titanic-Dataset.csv')

# -----------------------------
# Step 1: Exploration
# -----------------------------
print("=== Titanic Dataset Overview ===")
display(titanic_df.head())
display(titanic_df.info())
display(titanic_df.describe())

# -----------------------------
# Step 2: Data Cleaning
# -----------------------------
titanic_df = titanic_df.copy()  # Avoid chained assignment warnings
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].median())
titanic_df['Embarked'] = titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0])
# Check if 'Cabin' column exists before dropping it
if 'Cabin' in titanic_df.columns:
    titanic_df.drop('Cabin', axis=1, inplace=True)
titanic_df.drop_duplicates(inplace=True)

print("\nMissing values after cleaning:")
display(titanic_df.isnull().sum())

# -----------------------------
# Step 3: Data Analysis
# -----------------------------
# Survival by gender
survival_gender = titanic_df.groupby('Sex')['Survived'].mean() * 100
# Survival by class
survival_class = titanic_df.groupby('Pclass')['Survived'].mean() * 100
# Average age per class
avg_age_class = titanic_df.groupby('Pclass')['Age'].mean()
# Survival by age group
titanic_df['AgeGroup'] = pd.cut(titanic_df['Age'], bins=[0,12,18,60,100],
                                labels=['Child','Teenager','Adult','Senior'])
survival_age_group = titanic_df.groupby('AgeGroup')['Survived'].mean() * 100

print("\n=== Survival Rate by Gender (%) ===")
display(survival_gender)
print("\n=== Survival Rate by Class (%) ===")
display(survival_class)
print("\n=== Average Age per Class ===")
display(avg_age_class)
print("\n=== Survival Rate by Age Group (%) ===")
display(survival_age_group)

# -----------------------------
# Step 4: Filtering
# -----------------------------
female_survivors = titanic_df[(titanic_df['Sex']=='female') & (titanic_df['Survived']==1)]
child_survivors = titanic_df[(titanic_df['AgeGroup']=='Child') & (titanic_df['Survived']==1)]
first_class_survivors = titanic_df[(titanic_df['Pclass']==1) & (titanic_df['Survived']==1)]

print("\n=== Female Survivors ===")
display(female_survivors.head())
print(f"Total Female Survivors: {len(female_survivors)}")

print("\n=== Child Survivors ===")
display(child_survivors.head())
print(f"Total Child Survivors: {len(child_survivors)}")

print("\n=== 1st Class Survivors ===")
display(first_class_survivors.head())
print(f"Total 1st Class Survivors: {len(first_class_survivors)}")

# -----------------------------
# Step 5: Insights
# -----------------------------
print("\n=== Insights ===")
print(f"1. Gender: Women were more likely to survive ({survival_gender['female']:.2f}%) than men ({survival_gender['male']:.2f}%).")
print(f"2. Class: 1st class passengers had highest survival ({survival_class[1]:.2f}%), followed by 2nd ({survival_class[2]:.2f}%) and 3rd ({survival_class[3]:.2f}%).")
print(f"3. Children: Higher survival rate ({survival_age_group['Child']:.2f}%) compared to adults and seniors.")
print("4. Highest survival combination: Female passengers in 1st class had the highest likelihood of survival.")
```
