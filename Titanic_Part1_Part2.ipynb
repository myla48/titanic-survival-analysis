## Part 2: Real Dataset — Titanic

### Step 1: Exploration

```python
# Load the Titanic Dataset
titanic_df = pd.read_csv('/content/Titanic-Dataset.csv')

print("First 5 rows of the Titanic Dataset:")
display(titanic_df.head())

print("\nInformation about the Titanic Dataset:")
display(titanic_df.info())

print("\nDescriptive statistics of the Titanic Dataset:")
display(titanic_df.describe())
```

### Step 2: Data Cleaning

```python
# Handle missing values for 'Age' with the median
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)

# Handle missing values for 'Embarked' with the mode
# Get the most frequent value in the 'Embarked' column
most_frequent_embarked = titanic_df['Embarked'].mode()[0]
titanic_df['Embarked'].fillna(most_frequent_embarked, inplace=True)

# Drop the 'Cabin' column due to a large number of missing values
titanic_df.drop('Cabin', axis=1, inplace=True)

# Remove duplicates if any (though unlikely in this dataset for passenger IDs)
initial_rows = titanic_df.shape[0]
titanic_df.drop_duplicates(inplace=True)
duplicates_removed = initial_rows - titanic_df.shape[0]

print(f"Missing values after cleaning (Age, Embarked):")
display(titanic_df[['Age', 'Embarked']].isnull().sum())
print(f"'Cabin' column dropped. Remaining columns: {list(titanic_df.columns)}")
print(f"Number of duplicate rows removed: {duplicates_removed}")

print("\nInformation about the Titanic Dataset after cleaning:")
display(titanic_df.info())
```

### Step 3: Data Analysis

```python
# Survival rate by gender
survival_by_gender = titanic_df.groupby('Sex')['Survived'].mean() * 100
print("Survival Rate by Gender:")
display(survival_by_gender)

# Survival rate by class
survival_by_class = titanic_df.groupby('Pclass')['Survived'].mean() * 100
print("\nSurvival Rate by Class:")
display(survival_by_class)

# Average age per class
average_age_by_class = titanic_df.groupby('Pclass')['Age'].mean()
print("\nAverage Age per Class:")
display(average_age_by_class)

# Survival rate by age group
# Create age groups
titanic_df['AgeGroup'] = pd.cut(titanic_df['Age'], bins=[0, 12, 18, 60, 100], labels=['Child', 'Teenager', 'Adult', 'Senior'])
survival_by_age_group = titanic_df.groupby('AgeGroup')['Survived'].mean() * 100
print("\nSurvival Rate by Age Group:")
display(survival_by_age_group)
```

### Step 4: Filtering

```python
# Female passengers who survived
female_survivors = titanic_df[(titanic_df['Sex'] == 'female') & (titanic_df['Survived'] == 1)]
print("Female Passengers who Survived:")
display(female_survivors.head())
print(f"Total Female Survivors: {len(female_survivors)}")

# Children who survived (AgeGroup 'Child' and Survived == 1)
child_survivors = titanic_df[(titanic_df['AgeGroup'] == 'Child') & (titanic_df['Survived'] == 1)]
print("\nChildren who Survived:")
display(child_survivors.head())
print(f"Total Child Survivors: {len(child_survivors)}")

# Passengers in 1st class with high survival probability
# (This can be interpreted as 1st class passengers who survived)
first_class_survivors = titanic_df[(titanic_df['Pclass'] == 1) & (titanic_df['Survived'] == 1)]
print("\nPassengers in 1st Class who Survived:")
display(first_class_survivors.head())
print(f"Total 1st Class Survivors: {len(first_class_survivors)}")
```

### Step 5: Insights

Based on the analysis:

*   **Who was more likely to survive (gender)?**
    Women (`74.20%` survival rate) were significantly more likely to survive than men (`18.89%` survival rate).

*   **Did class affect survival?**
    Yes, class strongly affected survival. Passengers in 1st class had the highest survival rate (`62.96%`), followed by 2nd class (`47.28%`), and then 3rd class (`24.24%`). This indicates a clear advantage for higher classes.

*   **Were children prioritized?**
    Children (`57.97%` survival rate for 'Child' age group) had a higher survival rate compared to teenagers (`42.86%`), adults (`36.58%`), and seniors (`22.73%`). This suggests that children were indeed prioritized or had a better chance of survival.

*   **What combination had the highest survival rate?**
    Combining the insights, **female passengers in 1st class** would likely have had the highest survival rate. From the previous output, female survival was `74.20%` and 1st class survival was `62.96%`.
